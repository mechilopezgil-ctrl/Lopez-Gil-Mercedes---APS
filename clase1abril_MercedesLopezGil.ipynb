```python
'''Sea x[n]= 4 + 3 * sen(pi/2 * n), calcular la DFT X[k]'''

import numpy as np
import matplotlib.pyplot as plt

N = 8

n = np.arange(0, N, 1) #empiezo en 0, termino en N, pasos de 1

def x(n):
    return 4 + 3 * np.sin(np.pi/2 * n)

def X_transformada(k):
    return np.fft.fft(x(n))

X_t = X_transformada(n)

Mod_X = np.abs(X_t)
Pot_X = (Mod_X/N)**2 #calculo y normalizo el espectro de potencia

plt.figure()
plt.stem(Pot_X)
plt.title(label= 'Potencia de la transformada X[k]')
plt.grid(True)
plt.legend()
plt.show()

Fase_X = np.angle(X_t)

plt.figure()
plt.stem(Fase_X)
plt.title(label= 'Fase de la transformada X[k]')
plt.grid(True)
plt.legend()
plt.show()
```
