```python
import numpy as np
import matplotlib.pyplot as plt

vmax= 1
dc=0
f0=1
ph=0
n= 1000
fs= 1000
ts= 1/fs

def mi_funcion_sen ( vmax , dc , f0 , ph, n , fs):
    tt = np.linspace(0, (n-1)*ts, n)
    xx = dc + vmax * np.sin(2 * np.pi * f0 * tt + ph)
    return tt, xx


tt, xx = mi_funcion_sen( vmax , dc , f0 , ph, n , fs)

plt.plot(tt, xx)
plt.title("Onda Senoidal")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
sigma = 1
mu = 0
U_n = np.random.normal(mu, sigma, n)
xxn = xx + U_n

plt.plot(tt, xxn)
plt.title("Onda Senoidal con ruido")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
vmax = np.sqrt(2)

tt, xx = mi_funcion_sen(vmax, dc, f0, ph, n, fs)

Px = np.var(xx)
SNR = 20
Pr = 10**(-SNR/10)


U_n = np.random.normal(mu, np.sqrt(Pr), n)

xxn = xx + U_n


plt.figure(1)
plt.clf()
plt.title(f'f0 = {f0} Hz') 
plt.plot(tt, xxn)
plt.plot(tt, xx, 'r', lw=2)
```

```python
from scipy import signal as sig


n0 = 300
dd = np.zeros(n)
dd[n0] = 1

yy = sig.convolve(xx, dd)

plt.plot(yy)
```

```python
yy = 1/n*sig.convolve(U_n, np.flip(U_n))

plt.figure(1)
plt.clf()
plt.plot(yy)
```
